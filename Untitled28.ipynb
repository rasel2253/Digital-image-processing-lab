```python
nlp=spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    filtered_token = []
    doc = nlp(text)
    for token in doc:
        if token.is_stop or token.is_punct or token.is_space:
            continue
        filtered_token.append(token.lemma_)
    return " ".join(filtered_token)
```

```python
all_data['preprocessed_Answer']=all_data['Answer'].apply(preprocess)
```

```python
all_data.head()
```

```python
tfidf_vectorizer = TfidfVectorizer()

# Vectorize the "question" and "answer" columns
tfidf_questions = tfidf_vectorizer.fit_transform(all_data['Question'])
tfidf_answers = tfidf_vectorizer.fit_transform(all_data['preprocessed_Answer'])

```

#Training Model

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

from sklearn.pipeline import Pipeline
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
clf= Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
('Multi NB', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)

y_pred= clf.predict(x_test)
```

```python
print( accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf= Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
('Random Forest', RandomForestClassifier())
])
```

```python
clf.fit(x_train, y_train)

y_pred= clf.predict(x_test)
```
